```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the sample Superstore data into a DataFrame
data = {
    'State': ['California', 'Texas', 'New York', 'Florida', 'Illinois'],
    'Profit': [30000, 25000, 18000, 12000, 10000],
    'Sales': [150000, 120000, 90000, 60000, 50000],
    'Shipping_Cost': [5000, 4000, 3000, 2000, 1000],
    'Quantity': [500, 400, 350, 300, 250]
}

df = pd.DataFrame(data)

# Step 2: Profit Distribution Analysis - Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(df['Profit'], labels=df['State'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title('Profit Distribution by State')
plt.show()
```

```python
# Step 3: Profit-to-Sales Ratio Analysis
df['Profit_to_Sales_Ratio'] = df['Profit'] / df['Sales']
plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='Profit_to_Sales_Ratio', data=df, palette='Set1')
plt.title('Profit to Sales Ratio by State')
plt.xlabel('State')
plt.ylabel('Profit-to-Sales Ratio')
plt.show()
```

```python
# Step 4: Shipping Costs Visualization - Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='Shipping_Cost', data=df, palette='Set3')
plt.title('Shipping Costs by State')
plt.xlabel('State')
plt.ylabel('Total Shipping Cost')
plt.show()
```

```python
# Step 5: Sales vs Profit - Scatter plot to visualize the relationship between sales and profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', data=df, size='Quantity', sizes=(100, 400), hue='State', palette='Set2')
plt.title('Sales vs Profit by State')
plt.xlabel('Total Sales')
plt.ylabel('Total Profit')
plt.legend(title='State')
plt.show()
```

```python
# Selecting only numeric columns for the correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Generate the heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Superstore Metrics')
plt.show()
```
